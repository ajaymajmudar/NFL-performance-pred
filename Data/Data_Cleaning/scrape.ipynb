```python
year_range = range(1960, 2022)
url_base = 'https://www.pro-football-reference.com/years/'
year_dfs = []
for year in year_range:
    year_url = url_base + str(year) + "/draft.htm"
    year_df = pd.read_html(year_url)[0]
    year_df.columns = year_df.columns.droplevel()
    year_df["Year"] = year
    #print(year_df.columns)
    result_df = year_df[["Player", "Pos", "To", "CarAV", "DrAV","Age", "Tm", "Rnd", "Pick", "Year", "College/Univ"]]
    year_dfs.append(result_df)
draft_df = pd.concat(year_dfs, ignore_index=True)
draft_df.head()
```

```python
draft_df = draft_df[draft_df["Rnd"] != "Rnd"]
```

```python
draft_df.to_csv("draft_AV_table_60_21.csv")
```

```python
total_player_data = pd.read_csv("../PlayerDataset_2.csv")
total_player_data.shape
```

```python
total_player_with_av = pd.merge(total_player_data, draft_df, left_on = ["name", "position"], right_on = ["Player", "Pos"], how='left')
```

```python
total_player_with_av["Missing"] = total_player_with_av["Player"].isna()
total_player_with_av.head()
```

```python
players_used = total_player_with_av[total_player_with_av["Missing"] == False]

players_used = players_used[["Player", "Pos", "To", "CarAV", "DrAV", "Age", "Tm", "Rnd", "Pick", "Year", "College/Univ","Missing"]]
```

```python
players_missing = total_player_with_av[total_player_with_av["Missing"] == True]
players_missing.drop(["Player", "Pos", "To", "CarAV", "DrAV", "Age", "Tm", "Rnd", "Pick", "Year", "College/Univ", "Missing"], axis=1)
```

```python
draft_df["NewKey"] = draft_df["Player"] + draft_df["Pos"] + draft_df["CarAV"].astype(str) + draft_df["Year"].astype(str)
players_used["NewKey"] = players_used["Player"] + players_used["Pos"] + players_used["CarAV"].astype(str) + players_used["Year"].astype(int).astype(str)
```

```python
remaining = draft_df[~draft_df.NewKey.isin(players_used.NewKey)]
print(remaining.shape, draft_df.shape)
remaining
```
